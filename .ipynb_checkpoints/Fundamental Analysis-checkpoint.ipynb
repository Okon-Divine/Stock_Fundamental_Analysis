```python
import requests
import pandas as pd
import csv
import requests
pd.set_option('display.max_columns', None)
import matplotlib.pyplot as plt
```

#### Data Cleaning and Processing

```python
def call_api_(url):
    r = requests.get(url)
    data = r.json()
    return data

def get_stock_list(CSV_URL):
    data_list = []
    with requests.Session() as s:
        download = s.get(CSV_URL)
        decoded_content = download.content.decode('utf-8')
        cr = csv.reader(decoded_content.splitlines(), delimiter=',')
        my_list = list(cr)
        for row in my_list:
            data_list.append(row)
    return data_list


def reports(data, key):
    data_ = []
    for i, n in data.items():
        if i == key:
            data_.append(n)
    a = pd.DataFrame(data_[0])
    return a

def handle_data(a):
    for i in a.columns:
        null_array = []
        for n in a[i].values:
            if n == "None":
                null_array.append("0")
            else:
                null_array.append(n)
        a[i] = null_array
    return a

def handle_dtype(a):
    a['fiscalDateEnding'] = pd.to_datetime(a['fiscalDateEnding'])
    a_df = a.drop(['fiscalDateEnding','reportedCurrency'] , axis=1)
    for i in a_df.columns:
        a_df[i] = a_df[i].astype("float")
    obj_df = a[['fiscalDateEnding','reportedCurrency']]
    df_ = pd.concat([obj_df, a_df], axis=1)
    return df_
```

#### Define Variables

```python
ticker="Meta"
api_key= "##### Input yours --> https://www.alphavantage.co"
data_type_1 = "LISTING_STATUS"
data_type_2 = "INCOME_STATEMENT"
data_type_3 = "EARNINGS"
data_type_4 = "BALANCE_SHEET"
```

#### Get all stock listings (symbols, exchange, ipoDate, etc)

```python
data_list = get_stock_list(CSV_URL=f'https://www.alphavantage.co/query?function={data_type_1}&apikey={api_key}')
stock_list_ = pd.DataFrame(data_list)
stock_list_.columns = stock_list_.iloc[0].values
stock_list_ = stock_list_.drop(0, axis=0)
stock_list_.head()
```

#### Income statement data

```python
income_data = call_api_(url = f'https://www.alphavantage.co/query?function={data_type_2}&symbol={ticker}&apikey={api_key}')
df_annual_reports = reports(data=income_data, key='annualReports')
b = handle_data(a=df_annual_reports)
df_data_annual_report = handle_dtype(a=b)
df_data_annual_report
```

#### Calculate groos profit, operating margin, net profit margin

```python
df_data_annual_report["gross_profit_margin"] = (df_data_annual_report["grossProfit"]/df_data_annual_report["totalRevenue"])*100
df_data_annual_report["operating_Margin"] = (df_data_annual_report["operatingIncome"]/df_data_annual_report["totalRevenue"])*100
df_data_annual_report["Net_profit_Margin"] = (df_data_annual_report['netIncome']/df_data_annual_report["totalRevenue"])*100
df_data_annual_report
```

#### Balance Sheet

```python
bal_sheet_data = call_api_(url = f'https://www.alphavantage.co/query?function={data_type_4}&symbol={ticker}&apikey={api_key}')
df_annual_reports = reports(data=bal_sheet_data, key='annualReports')
b = handle_data(a=df_annual_reports)
df_data_annual_report_b = handle_dtype(a=b)
df_data_annual_report_b
```

#### Earnings

```python
earn_sheet_data = call_api_(url = f'https://www.alphavantage.co/query?function={data_type_3}&symbol={ticker}&apikey={api_key}')

def earn_data_sheet(time_earn):
    for i, n in earn_sheet_data.items():
        if i == time_earn:
            return n
 
df_earn = earn_data_sheet(time_earn='annualEarnings')
eps_annual = pd.DataFrame(df_earn)

eps_annual["fiscalDateEnding"] = pd.to_datetime(eps_annual["fiscalDateEnding"])
eps_annual["reportedEPS"] = pd.to_numeric(eps_annual["fiscalDateEnding"])

eps_annual.head()
```

#### Choose eps range from 2018 to 2022

```python
eps_range = eps_annual[(eps_annual.fiscalDateEnding >= "2018-12-31") & (eps_annual.fiscalDateEnding <= "2022-12-31")]
```

#### Merge all data (balance sheet, earnings, income statement) together for that year range 

```python
df_eps = pd.merge(df_data_annual_report_b,eps_range, on="fiscalDateEnding").drop("reportedCurrency", axis=1)
new_df = pd.merge(df_data_annual_report,df_eps, on="fiscalDateEnding")
new_df
```

#### Calculate ROE (return on equity)

```python
new_df["ROE"] = new_df["netIncome"]/new_df["totalShareholderEquity"]
```

#### Select all the variables(columns for analysis)

```python
cols = ["fiscalDateEnding", "gross_profit_margin", 
        "operating_Margin", "Net_profit_Margin", "reportedEPS", "ROE" ]

df_ = new_df[cols]
df_
```

#### Visualizations

```python
plt.style.use('seaborn')
fig, axs = plt.subplots(nrows=3, ncols=2, figsize=(12, 8))
axs = axs.ravel()

for i, col in enumerate(df_.columns[1:]):
    axs[i].plot(df_['fiscalDateEnding'], df_[col], color='C{}'.format(i), linewidth=2)
    axs[i].set_title(col)
    axs[i].set_xlabel('fiscalDateEnding')
    axs[i].set_ylabel(col)
    axs[i].grid(True)
    axs[i].legend([col])

fig.tight_layout()

plt.show()
```
